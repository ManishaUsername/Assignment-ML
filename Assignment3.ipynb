```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\hp\\OneDrive\\Desktop\\Assignmwnt ML\\Assignmwnt ML\\Customers - Customers.csv")
```

```python
print("Head of dataframe: ")
print(df.head())
```

```python
print(df.info())
```

```python
# Scatter Plot

plt.figure(figsize=(10, 6))
plt.scatter(x='Annual Income ($)', y='Spending Score (1-100)', data=df, color='Green')
plt.title('Annual Icome VS Spending Score')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
```

```python
# Bar Plot

plt.figure(figsize=(15,6))

sns.barplot(x='Age', y='Annual Income ($)', data=df, color='Pink')

plt.xticks(rotation=90)
plt.title('Age VS Annaul Income')

plt.show()
     
```

```python
# Box Plot

plt.figure(figsize=(10, 6))
sns.boxplot(x='Profession', y='Spending Score (1-100)', data=df, color='Skyblue')
plt.xticks(rotation=45)
plt.title('Profession VS Spending Score')
plt.show()
```

```python


# Assuming 'df' is your DataFrame and it contains a 'Gender' column
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=df, palette="coolwarm")
plt.title("Gender (0 = Female, 1 = Male)", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python


# Assuming 'df' is your DataFrame
family_size_counts = df['Profession'].value_counts()
plt.figure(figsize=(8, 8))
family_size_counts.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90, explode=[0.02]*len(family_size_counts))
plt.title('Profession Distribution')
# plt.legend()
plt.show()
```

```python
# Line Plot

data_sorted = df.sort_values(by='Age')

plt.figure(figsize=(10, 6))
plt.plot(data_sorted['Age'], data_sorted['Spending Score (1-100)'], marker='o', linestyle='-', color='Skyblue'

, label='Spending Score')
plt.title('Spending Score by Age')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.grid()
plt.show()
     
```
